```python
import pandas as pd
import urllib
from sqlalchemy import create_engine

def load_mdb_data(file_path):
    try:
        # Encoded file path for the connection string
        encoded_file_path = urllib.parse.quote_plus(f'DRIVER={{Microsoft Access Driver (*.mdb, *.accdb)}};DBQ={file_path};')
        
        # Create SQLAlchemy engine
        engine = create_engine(f'mdb+access:///?odbc_connect={encoded_file_path}')

        # Retrieving all table names in the MDB file
        with engine.connect() as conn:
            tables = engine.table_names()
            
            # Reading each table into a pandas DataFrame
            dataframes = {table: pd.read_sql_table(table, conn) for table in tables}

        return dataframes
    except Exception as e:
        print(f"Error occurred: {e}")
        return None

# Example usage
file_path = "/Users/Jean-BaptistePERNEY/Documents/ECHO/DATA/medical_database_2022/01022022.mdb"
dataframes = load_mdb_data(file_path)

if dataframes:
    for table_name, df in dataframes.items():
        print(f"Table: {table_name}")
        print(df.head())
```

```python
import pyodbc
import pandas as pd
import os

def load_mdb_data(file_path):
    conn_str = (
        r'DRIVER={Microsoft Access Driver (*.mdb, *.accdb)};'
        r'DBQ=' + file_path + ';'
    )
    conn = pyodbc.connect(conn_str)
    cursor = conn.cursor()
    table_names = [table.table_name for table in cursor.tables(tableType='TABLE')]
    
    dataframes = {}
    for table_name in table_names:
        query = f"SELECT * FROM [{table_name}]"
        df = pd.read_sql(query, conn)
        df['Source'] = os.path.basename(file_path)  # Ajouter la colonne Source ici
        dataframes[table_name] = df

    cursor.close()
    conn.close()

    return dataframes

directory_path = "/Users/Jean-BaptistePERNEY/Documents/ECHO/DATA/telecom_log_2022"
all_dataframes = {}

for file_name in os.listdir(directory_path):
    if file_name.endswith('.mdb'):
        file_path = os.path.join(directory_path, file_name)
        dataframes = load_mdb_data(file_path)

        for table_name, df in dataframes.items():
            if table_name in all_dataframes:
                # Concaténer par lignes si la table existe déjà
                all_dataframes[table_name] = pd.concat([all_dataframes[table_name], df], axis=0)
            else:
                # Sinon, ajouter la table au dictionnaire
                all_dataframes[table_name] = df

# Afficher les premières lignes de chaque DataFrame du dictionnaire
for table_name, df in all_dataframes.items():
    print(f"Table: {table_name}")
    print(df.head())
```
